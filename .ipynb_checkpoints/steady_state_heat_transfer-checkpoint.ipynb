```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
```

```python
start = timeit.default_timer()

temp = [40, 40, 0, 0]

tot_mesh_dim = 20
inner_mesh_dim = tot_mesh_dim - 2
mesh_elems = inner_mesh_dim**2

# defining the plate
plate = np.zeros((tot_mesh_dim, tot_mesh_dim), np.uint8)
plate[0,:] = mesh_elems + 1
plate[:,-1] = mesh_elems + 2
plate[-1,:] = mesh_elems + 3
plate[:,0] = mesh_elems + 4

node_num = 1
for i in range(tot_mesh_dim-2, 0, -1):
    for j in range(1, tot_mesh_dim-1):
        plate[i,j] = node_num
        node_num += 1

print(plate)
```

```python
A_lg = np.zeros((mesh_elems, mesh_elems+4))

# Define A_lg as A with b's properties built in
for i in range(0,tot_mesh_dim):
    for j in range(0,tot_mesh_dim):
        if plate[i,j] <= mesh_elems:
            # current
            A_lg[plate[i,j]-1, plate[i,j]-1] = -4
            # bottom
            A_lg[plate[i,j]-1, plate[i+1,j]-1] = 1
            # top
            A_lg[plate[i,j]-1, plate[i-1,j]-1] = 1
            # right
            A_lg[plate[i,j]-1, plate[i,j+1]-1] = 1
            # left
            A_lg[plate[i,j]-1, plate[i,j-1]-1] = 1
            
print(A_lg)
```

```python
# Split up A_lg
b_matrix = A_lg[:, -4:]
A = A_lg[:, :mesh_elems]

# define b from b_matrix
b = np.zeros((mesh_elems, 1))
for i in range(0, mesh_elems):
    for j in range(0, 4):
        if b_matrix[i,j]==1:
            b[i] -= temp[j]
print(b)
```

```python
x = np.linalg.solve(A, b)
T = np.reshape(x, (inner_mesh_dim, inner_mesh_dim))
plt.contour(T)
plt.show


stop = timeit.default_timer()

print('Time: ', stop - start)
```
